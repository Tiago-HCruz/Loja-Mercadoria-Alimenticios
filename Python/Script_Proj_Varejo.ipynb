### Função que retorna um vetor com datas simuladas

A Função foi feita para retorna datas aleatorias do ano de 2025

```python
## Função que retorna data aleatoria com base em um ano definido (nesse caso, 2025)
def Data(Ano, Tamanho):
  Data = pd.DataFrame({
      "Mes": [random.randrange(1, 13) for x in range(Tamanho)],
      "Ano": Ano
      })
  def Dia(variavel):
        if variavel == 1 or variavel == 3 or variavel == 5 or \
           variavel == 7 or variavel == 8 or variavel == 10 or \
           variavel == 12:
           Dias = random.randrange(1, 32)
        else:
           if variavel == 2:
              Dias = random.randrange(1, 29)
           else:
              Dias = random.randrange(1, 31)
        return Dias
  Data["Dia"] = Data.loc[:, ["Mes"]].\
                    apply(lambda x: Dia(x['Mes']),
                          axis=1)
  Data = ['/'.join([str(Data.loc[x,"Dia"]), str(Data.loc[x,"Mes"]), Ano]) for x in range(len(Data))]
  return Data
```
